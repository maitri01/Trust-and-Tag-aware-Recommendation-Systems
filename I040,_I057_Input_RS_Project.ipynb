##User-User

```python
import numpy as np
import pandas as pd
```

```python
user_friends = pd.read_csv('/content/hetrec2011-lastfm-2k/user_friends.csv')
user_friends
```

```python
# Create a pivot table with index starting at zero and ending at 2100
pivot_table = user_friends.pivot_table(index=['userID'], columns='friendID', aggfunc='size', fill_value=0)

# Reindex the pivot table to have index from 0 to 2100
# user_friends_df = pivot_table.reindex(range(1, 2101), fill_value=0, axis=0).reindex(range(1, 2101), fill_value=0, axis=1)
user_friends_df = pivot_table

# Reset column names to be 'userID' and 'friendID'
user_friends_df.columns.name = 'friendID'
user_friends_df.index.name = 'userID'

print(user_friends_df)

matrix =user_friends_df.to_numpy()
matrix
```

```python
useruser_dict = {i+1: index for i, index in enumerate(user_friends_df.index)}
```

```python
useruser_dict
```

```python
matrix.shape
```

```python
df = pd.DataFrame(matrix)
df
```

```python
df.to_csv('user_user.csv', index=False)
```
